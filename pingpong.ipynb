```python
#imports

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load csv file
df = pd.read_csv("C:\\Users\\green\\OneDrive\\Desktop\\cat.csv")

# print first 5 rows
print(df.head())
```

```python
# graph the data
plt.scatter(df['MESSAGE SIZE'], df['TOTAL TIME'])
plt.xlabel('Message Size (bytes)')
plt.ylabel('Total Time (s)')

# show the graph
plt.show()
```

```python
# calculate slope (bandwidth) and intercept (latency)
slope, intercept = np.polyfit(df['MESSAGE SIZE'], df['TOTAL TIME'], 1)
print(f"Bandwidth: {slope} bytes/s")
print(f"Latency: {intercept} s")

# calculate the R^2 value
r_squared = np.corrcoef(df['MESSAGE SIZE'], df['TOTAL TIME'])[0, 1]**2
print(f"R^2: {r_squared}")
```
